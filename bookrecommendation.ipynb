```python
ratings__with_name = ratings.merge(books, on ='ISBN')
```

```python
num_rating_df=ratings__with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating' : 'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings__with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating' : 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_ratings = avg_rating_df.merge(num_rating_df, on ='Book-Title')
popular_ratings
```

```python
popular_ratings=popular_ratings[popular_ratings['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(10)
```
